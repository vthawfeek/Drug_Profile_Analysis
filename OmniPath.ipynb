```python
# Functions

#Function to obtain the first-order neighborhood around an input protein, and the net regulatory effect
def first_order(name, R):

    #First order causally afected proteins
    a = list(pa.up_affects(name).up())

    #list to keep track of net effect on the neighborhood proteins
    reg = [R] * len(a)

    #Change net regulation based on sign of interaction in the edge.
    for index, protein in enumerate(a):
        edge = pa.get_edge(name, protein)
        d_edge = edge['dirs']
        sign = d_edge.consensus_edges()
        if sign[-1][-1] == 'negative':
            reg[index] = reg[index]*-1
        elif sign[-1][-1] != 'positive':
            reg[index] = 0

    #return output
    return a,reg
```

#### Higher Order 

```python
# Import libraries

import numpy as np
from pypath.legacy import main
import csv
import pandas as pd
```

```python
# Functions


#Recursive path constructor function. Returns list of paths (each as a list of uniprot IDs) between
#the input start and end node.
def find_all_paths(start, end, maxlen, path=[]):
    path = path + [start]
    if start == end:
        return [path]

    paths = []
    if maxlen is None or len(path) <= maxlen:
        setB = set(pa.up_affects(start).up())
        intersect = set(path) & setB
        setA = setB - intersect
        for node in setA:
            paths.extend(find_all_paths(node, end, maxlen, path))

    return paths



#Function for finding the list of transcription factors within given order network about a single protein,
#and an array counting the number of net 'activation', 'inhibition' and 'unknown effect' paths
#between protein and transcrition factor.
def tf_finder(protein, R, max_order, max_len):

    #produce the directed graph around each protein in direct-drug-target list
    sub_network = pa.get_directed(
        pa.graph.induced_subgraph(
            list(pa.up_neighborhood(protein, order = max_order).vs()
        )
    ))

    #Case of no directional information about this protein
    if sub_network is None:
        return [], None
    
    else:

        #list of proteins that are to transcritpion factors
        v_tfs=[v['name'] for v in sub_network.vs() if v['tf']]
        ind_to_delete = []

        #First column keeps track of count of net activation paths, second of net inhibition,
        #third of net unknown effect to the transcription factor.
        regCountArray = np.zeros((len(v_tfs),3))
        
        for h,v in enumerate(v_tfs):
            j = find_all_paths(protein, v, max_len)

            if len(j) == 0:
                ind_to_delete = ind_to_delete + [h]

            else:
                for i in range(len(j)):         #select a path
                    x=R
                    for k in range(len(j[i])-1):        #for all nodes in selected path
                        edge = pa.get_edge(j[i][k], j[i][k+1])['dirs']
                        sign = edge.consensus_edges()[-1][-1]
                        if sign == 'negative':
                            x = x * (-1)
                        elif sign != 'positive':
                            x = 0

                    if x > 0:                           #Add path counts
                        regCountArray[h,0]=regCountArray[h,0]+1
                    elif x < 0:
                        regCountArray[h,1]=regCountArray[h,1]+1
                    else:
                        regCountArray[h,2]=regCountArray[h,2]+1


        v_tfs = np.delete(v_tfs, ind_to_delete).tolist()    #delete transcription factors already present from previous proteins
        regCountArray = np.delete(regCountArray, ind_to_delete, axis=0)

                    
        return v_tfs, regCountArray

    
    
#This function constructs the data frame containing the list of all transcription factors prdicted to 
#be affected by the drug, and the counts of net activation, inhibition, and unknown effect paths to these transcription factors
def drug_tf_profile(zeroth_list, zeroth_regulator_list, max_order, max_len):
    
    #initiate TF list and regulator array from zeroth protein
    TFs, regulator_array = tf_finder(zeroth_list[0],zeroth_regulator_list[0], max_order, max_len)

    #For each direct target of drug
    for i , protein in enumerate(zeroth_list[1:]):

        #output the list of transcription factors downstream in its causal network, and net regulation effect
        TFtemp, regulator_array_temp = tf_finder(protein, zeroth_regulator_list[i+1],max_order, max_len)

        if regulator_array_temp is not None:    #In the case of no paths to tfs
            if regulator_array is not None:
                
                #Here finding the transcription factors already in the (growing list) and adding
                #to their activation/inhibition/unknown paths count (to have final unique list)
                common_TFs_index= [j for j, name in enumerate(TFtemp) if name in TFs]
                common_TFs = [name for j, name in enumerate(TFtemp) if name in TFs]
                if len(common_TFs) != 0:
                    for k , common_TF in enumerate(common_TFs):
                        ind = TFs.index(common_TF)
                        if np.shape(regulator_array)[0] != 0 :
                            regulator_array[ind, 0:2] = regulator_array[ind, 0:2] + regulator_array_temp[common_TFs_index[k], 0:2]

                TF_new = np.delete(TFtemp, common_TFs_index, 0)
                TF_new = TF_new.tolist()
                regulator_array_new = np.delete(regulator_array_temp, common_TFs_index, 0)

                TFs = TFs+ TF_new
                regulator_array = np.append(regulator_array, regulator_array_new, axis=0)

            else:
                TFs= TFtemp			#Initiate array if zeroth protein is none
                regulator_array = regulator_array_temp


    if regulator_array is None:
        print('No transcription factors found in network')
        return

    else:

        reg0 = regulator_array[:,0].tolist()
        reg1 = regulator_array[:,1].tolist()
        reg2 = regulator_array[:,2].tolist()

    
    #Data as a pandas data frame
        data = {'Transcription factors' : TFs,
                'Activation path counts' : reg0,
                'Inhibition path counts' : reg1,
                'Undetermined effect path count' : reg2}

        df=pd.DataFrame(data)

        print(df)

        return df
```

```python
DrugTarget_dict={'Methamphetamine':{'P14416':1,'P35368':-1},'Drug5FU':{'P42858':1,'P16473':-1,'Q01453':1,'P10636':1,'P40225':-1,'Q16637':1,'P02545':-1,'P04818':-1,'Q194T2':1}}
Drugs=list(DrugTarget_dict.keys())
Tissues=['Lung','Liver','Blood','Kidney','Spleen']
Network_df=edges_df
Target_Degree=0
Top_tf_Counts=10
Tissue=Tissues

# set path
notebook_path = os.path.abspath("OmniPath.ipynb")
path=notebook_path.rsplit('/',1)[0]+'/'

# read ChEMBL drug list
drug_data = pd.read_excel(path+'All_ChEMBL_Drugs.xlsx')
drug_data = drug_data[['molecule_chembl_id','Entry']].rename(columns={'molecule_chembl_id':'id_a','Entry':'id_b'})
drug_data['directed']=True
drug_data['effect']=-1
Network_df = pd.concat((drug_data,Network_df),axis=0,ignore_index=True)
Network_df['edge'] = Network_df.id_a.astype(str) + '-->' + Network_df.id_b.astype(str)

## node table
nodes=set(list(Network_df['id_a'])+list(Network_df['id_b']))
nodes_df=pd.DataFrame(nodes,columns=['UniProtID'])

# read tissue expression data
hpa = pd.read_table(path+'normal_tissue.tsv')
hpa.rename(columns={'Gene':'ensGene'},inplace=True)
uni_ens_map = pd.read_table(path+'UniProt_Ens_Map.tsv')
uni_ens_map.dropna(inplace=True)
##split Ensembl ID maps
co = uni_ens_map.shape[0]
for label,row in uni_ens_map.iterrows():
    ens = row['ensGene'].split(',')
    if len(ens)>1:
        for i in ens:
            uni_ens_map.loc[co]=[row['#acc'],i]
            co+=1
## remove multi-Ensembl ID maps
for label,row in uni_ens_map.iterrows():
    ens = row['ensGene'].split(',')
    if len(ens)>1:
        uni_ens_map.drop(label,inplace=True)
        
hpa_mapped = hpa.merge(uni_ens_map,how='inner', on='ensGene')
hpa_mapped.rename(columns={'#acc':'UniProtID'},inplace=True)
hpa_mapped = hpa_mapped[hpa_mapped['Reliability']=='Approved']

## merge with node data
nodes_df = nodes_df.merge(hpa_mapped[['UniProtID','Tissue','Level']],how='left',on='UniProtID')
nodes_df.drop_duplicates(inplace=True)

## tissue column values to individual columns
nodes_df = nodes_df.pivot_table(values='Level',index='UniProtID',columns='Tissue',aggfunc='first')
nodes_df.reset_index(inplace=True)

## set [nan,not detected and low]=0 and [medium,high]=1
nodes_df.replace(to_replace=['Low','Not detected',np.nan],value=0,inplace=True)
nodes_df.replace(to_replace=['Medium','High'],value=1,inplace=True)

# set tissue
tissue = 'kidney'
exp_dict = nodes_df.set_index('UniProtID')[tissue].to_dict()
edge_dict = Network_df.set_index('edge')['effect'].to_dict()

# iterate over all drugs
frames = []
c = 0
for drug in list(drug_data['id_a'].unique()):

    # build network
    G=nx.MultiDiGraph()
    color_dict={'Drug':'red','target':'orange','1st_degree':'gold','2nd_degree':'pink','tf':'lime','other':'white',1:'blue',-1:'red'}
    all_ids = set(list(Network_df['id_a'])+list(Network_df['id_b'])+list(nodes_df['UniProtID']))

    for label,row in nodes_df.iterrows():
        G.add_node(row['UniProtID'])
    for label,row in Network_df.iterrows():
        G.add_edge(row['id_a'],row['id_b'])


    #tfs_in_graph = tfs_in_network & set([Val2Key(nodes_dict,i) for i in list(G.nodes)])
    tfs_in_graph = tfs_in_network & set(G.nodes)
    deg_dict={i:G.degree[i] for i in tfs_in_graph}
    deg_sort=sorted(deg_dict.values(),reverse=True)
    Top_tfs=set([Val2Key(deg_dict,i) for i in deg_sort])#[0:t]])
    Short_Path=[nx.shortest_path(G,source=drug,target=i) for i in Top_tfs if nx.has_path(G,drug,i)]

    # calculate effect of drug on tf
    score_list = []
    for p in range(len(Short_Path)-1):
        effect_values = []
        for i in range(len(Short_Path[p])-1):
            # set initial value based on expression
            if Short_Path[p][i+1] in exp_dict:
                iv = exp_dict[Short_Path[p][i+1]]
            else:
                iv = 0
            iv = 1 # assume all proteins are expressed
            # set sign based on interaction type
            lin = Short_Path[p][i]+'-->'+Short_Path[p][i+1]
            if lin in edge_dict:
                sign = edge_dict[lin]
            else:
                sign = 1
            # calculate score
            effect_values.append(iv*sign)
        tf_value = np.prod(effect_values)
        score_list.append([Short_Path[p][0],Short_Path[p][len(Short_Path[p])-1],tf_value])

    score_df = pd.DataFrame(score_list,columns=['Drug','TF','Effect'])
    frames.append(score_df)
    print(f"{c}/{len(list(drug_data['id_a'].unique()))}")
    c+=1
    
# create table
all_scores_df = pd.concat(frames)
score_table = all_scores_df.pivot_table(values='Effect',index='Drug',columns='TF',aggfunc='first')

# write output to file
score_table.to_csv(path+'Score_Table.csv',index=True,header=True)
```
